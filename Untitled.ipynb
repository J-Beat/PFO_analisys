```python
import geopandas
dfg = geopandas.read_file(g('*3.9.geojson')[0])
dfg['month'] = dfg['day'].apply(lambda x: x[0:7])
dfg = dfg.rename(columns=({'fig_number':'id'}))
dfg.head()
```

```python
geo1 = dfg[['id', 'type', 'geometry']].drop_duplicates('id')
data1 = dfg.groupby(['id', dfg.columns[3], 'month'], as_index=False).agg({'totalSum_mean':'mean', 'receiptCount_mean':'mean'})
```

```python
from ipywidgets import interact
from ipywidgets import widgets
import folium
import pandas as pd

schemes = ["normal.day", "normal.night"]


from ipywidgets import widgets
column = list(dfg[dfg.columns[3]].unique())
key = [data1.columns[3], data1.columns[4]]
date = list(dfg['month'].unique())

@interact(column=column, key=key, date=date)
def get_here_maptile(column=column[0], key=key[0], date=date[0]):
    data = data1[(data1['month']==date)&(data1[data1.columns[1]]==column)]
    data = data.rename(columns={'id':'fig'})
    data = data[['fig', key]]
    geo = geo1[geo1['type']==column]
    geo = geo[['id', 'geometry']]
    geo.crs = {'init': 'epsg:4326'}
#     print(data.head())
#     print(geo.head())
    b = list(data[key].quantile([0, 0.25, 0.5, 0.75, 1]))
    b[0] = b[0]-1
    b[-1] = b[-1]+1
    bins=[]
    for i in b:
        i = round(i)
        bins.append(i)
    print(bins, data[key].max(), data[key].min())
    m = folium.Map(location=[55.795129, 49.109482], zoom_start=5.5)

    folium.Choropleth(

        geo_data=geo,
        name='choropleth',
        data=data,
        columns=['fig', key],
        key_on='feature.properties.id',
        fill_color='BuPu',
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name='mean',
        bins = bins,
        reset=True
    ).add_to(m)

    folium.LayerControl().add_to(m)

    return m
```

```python
import sys
%matplotlib notebook
from matplotlib import pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
from ipywidgets import IntProgress
import bisect
import collections
import geopandas as gpd
import descartes
import folium
from geopy import distance 

gt = gpd.read_file('admin_level_3.geojson')
gt = gt[['name','geometry']]
gt.columns = ['Region','geometry']
gt = gt[gt['Region'] == 'Приволжский федеральный округ']

gpfo = gpd.read_file('admin_level_4.geojson')
gpfo = gpfo[['name','geometry']]
gpfo.columns = ['region_name','geometry']
gpfo = gpd.sjoin(gt, gpfo, how="right", op='contains')
gpfo = gpfo[~gpfo['Region'].isna()]
gpfo = gpfo[['region_name', 'geometry']]

go = gpd.read_file('admin_level_6.geojson')
go = go[['name','geometry']]
go.columns = ['district','geometry']
go = gpd.sjoin(gt, go, how="right", op='contains')
go = go[~go['Region'].isna()].reset_index()
go = go[['Region', 'district', 'geometry']]
go = gpd.sjoin(gpfo, go, how="right", op='contains').reset_index()
go = go[['region_name', 'district', 'geometry']]
go = go.loc[go['district'].str.contains(r'район|округ')]

```

```python
go.head()
```

```python
dff = pd.read_csv(g('PFO*type*')[0], sep = ',', dtype=str)
dff = dff.astype({'Сумма общая':float, 'Общее кол-во чеков':float})
dff['month'] = dff['Дата'].apply(lambda x: x[:7])
dff = dff.rename(columns={'Наименование регоина':'region_name', 'Административный округ':'district', 'Дата':'date'})
dff = dff.groupby(['region_name', 'district', 'type', 'month'], as_index=False).agg({'Сумма общая':sum, 'Общее кол-во чеков':sum})
dff['Сумма общая (млн)'] = dff['Сумма общая']/1000000
dff['Общее кол-во чеков (тыс)'] = dff['Общее кол-во чеков']/1000
# dff = dff.merge(go, on='district', how='left')
dff.head()
```

```python
from ipywidgets import interact
from ipywidgets import widgets
import folium
import pandas as pd

from ipywidgets import widgets
column = list(dff[dff.columns[2]].unique())
key = [dff.columns[6], dff.columns[7]]
date = list(dff['month'].unique())
region = list(dff['region_name'].unique())


@interact(column=column, key=key, date=date, region=region)
def get_here_maptile(column=column[0], key=key[0], date=date[0], region = region[0]):
    data = dff[(dff['month']==date)&(dff[dff.columns[2]]==column)&(dff['region_name']==region)]
#     data = data.rename(columns={'id':'fig'})
    data = data[['district', key]]
    print(len(data))
    geo = go[go['region_name']==region]
    geo = geo.reset_index()#rename(columns={'region_name':'id'})
    geo = geo[['district', 'geometry']]
    geo.crs = {'init': 'epsg:4326'}
#     print(data.head(), geo.head())
    b = list(data[key].quantile([0, 0.25, 0.5, 0.75, 1]))
    print(b)
    b[0] = b[0]-1
    b[-1] = b[-1]+1
    bins=[]
    for i in b:
        i = round(i)
        bins.append(i)
    print(bins, data[key].max(), data[key].min())
    m = folium.Map(location=[55.795129, 49.109482], zoom_start=5.5)

    folium.Choropleth(

        geo_data=geo,
        name='choropleth',
        data=data,
        columns=['district', key],
        key_on='feature.properties.district',
        fill_color='YlGn',
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name=key,
#         bins = bins,
#         reset=True
    ).add_to(m)

    folium.LayerControl().add_to(m)

    return m
```
